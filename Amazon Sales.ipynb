# Продажи с сайта амазон

Данные взяты с сайта kaggle.com: https://www.kaggle.com/datasets/karkavelrajaj/amazon-sales-dataset


## Датасет содержит в себе данные рейтингов и обзоров более тысячи продуктов Амазон в соответствии с информацией, указанной на официальном сайте Амазон.

Описание данных:

- `product_id` - айди продукта
- `product_name` - наименование продукта
- `category` - категория
- `discounted_price` - цена со скидкой
- `actual_price` - регулярная цена
- `discount_percentage` - процент скидки
- `rating` - рейтинг
- `rating_count` - количество проголосовавших людей
- `about_product` - описание продукта
- `user_id` - айди польщователя, написавшего обзор 
- `user_name` - имя пользователя, написавшего обзов 
- `review_id` - айди ревью
- `review_title` - короткая версия ревью
- `review_content` - полная версия ревью
- `img_link` - ссылка на картинку продукта
- `product_link` - ссылка на продукт на сайте амазон

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\datasets\\amazon.csv')
```

```python
df.head()
```

```python
#сразу дропнем ненужные нам столбцы
df = df.drop(columns=['review_title', 'review_content', 'img_link', 'product_link'])
```

```python
df = df.drop(index=[282, 324])
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df.discounted_price = df.discounted_price.str.replace('₹', '') 
df.discounted_price = df.discounted_price.str.replace(',', '') 
df.actual_price = df.actual_price.str.replace('₹', '')
df.actual_price = df.actual_price.str.replace(',', '')
df.discount_percentage = df.discount_percentage.str.replace('%', '')
df.rating_count = df.rating_count.str.replace(',' , '')
df.rating = df.rating.str.replace(',' , '')
```

```python
def percents(value):
    value /= 100
    return value
```

```python
df.info()
```

```python
df.head(5)
```

```python
df.rating.value_counts()
```

```python
df.loc[df.rating == '|']
```

```python
df.drop(index=1279, inplace=True)
```

```python
df.discounted_price = df.discounted_price.astype('float64')
df.actual_price =  df.actual_price.astype('float64')
df.discount_percentage =df.discount_percentage.astype('float64')
df.rating_count = df.rating_count.astype('float64')
df.rating = df.rating.astype('float64')
```

```python
df['discount_percentage'] = df['discount_percentage'].apply(percents)
```
